### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Esta funcion imprime los nombres de las columnas en la dataFrame

tienda.columns
```

```python
#Esta funcion imprime  la informacion completa de cada tienda y de los tipos de datos

print("TIENDA 1")
display(tienda.head())
print(tienda.info())
print(tienda.describe(include='all'))
print(tienda.isna().sum())

print("\nTIENDA 2")
display(tienda2.head())
print(tienda2.info())
print(tienda2.describe(include='all'))
print(tienda2.isna().sum())

print("\nTIENDA 3")
display(tienda3.head())
print(tienda3.info())
print(tienda3.describe(include='all'))
print(tienda3.isna().sum())

print("\nTIENDA 4")
display(tienda4.head())
print(tienda4.info())
print(tienda4.describe(include='all'))
print(tienda4.isna().sum())
```

```python
#Aqui se cra una nueva columna llamada "ingreso" en cada tienda en donde contemplamos el precio unicamente.

tienda['Ingreso'] = tienda['Precio']
tienda2['Ingreso'] = tienda2['Precio']
tienda3['Ingreso'] = tienda3['Precio']
tienda4['Ingreso'] = tienda4['Precio']
```

```python
#Sumamos el total de ingresos [por cada tienda e imprimimos el resultado

ingresos_t1 = tienda['Ingreso'].sum()
ingresos_t2 = tienda2['Ingreso'].sum()
ingresos_t3 = tienda3['Ingreso'].sum()
ingresos_t4 = tienda4['Ingreso'].sum()

print(ingresos_t1, ingresos_t2, ingresos_t3, ingresos_t4)
```

```python
#Lo presentamos en un dataFrame comparativo

import pandas as pd

ingresos = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ingresos Totales": [ingresos_t1, ingresos_t2, ingresos_t3, ingresos_t4]
})

ingresos
```

```python
#Importamos el grafico que representa la informacion previa

import matplotlib.pyplot as plt

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos_valores = [ingresos_t1, ingresos_t2, ingresos_t3, ingresos_t4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos_valores)
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingresos Totales")
plt.show()
```

📄 INFORME FINAL – Análisis de Tiendas del Sr. Juan
Introducción

El objetivo de este análisis fue determinar en cuál de las cuatro tiendas (Tienda 1, Tienda 2, Tienda 3 y Tienda 4) debería vender sus productos el Sr. Juan. Para ello, se evaluaron distintos indicadores clave basados en los conjuntos de datos proporcionados:

Ingresos totales por tienda

Categorías de productos más y menos vendidas

Calificaciones promedio otorgadas por los clientes

Productos más y menos vendidos

Costo de envío promedio (pagado por el cliente)

A partir de estos parámetros se elaboraron gráficos, mediciones estadísticas y comparaciones entre tiendas para identificar cuál ofrece las mejores condiciones comerciales.

Desarrollo del análisis
1. Ingresos totales por tienda

Los ingresos totales muestran la fuerza de ventas general de cada tienda.
El análisis reveló que:

Tienda 1 obtuvo los ingresos más altos.

Le siguen Tienda 2, Tienda 3 y finalmente Tienda 4.

Esto indica que la Tienda 1 posee mayor flujo comercial, lo que puede corresponder a mayor tráfico de clientes y mejores oportunidades de venta.

2. Cantidad de productos vendidos por categoría

Al agrupar por categoría y contar ventas por tienda, se observó que:

Todas las tiendas tienen ciertas categorías significativamente más vendidas, lo que indica preferencia del cliente.

Las categorías más populares se mantienen consistentes en todas las tiendas, pero la Tienda 1 y Tienda 2 presentan mayor volumen de ventas, confirmando que tienen mayor movimiento.

Esto es positivo para el Sr. Juan, porque tiendas con más demanda tienden a rotar más productos externos.

3. Calificación promedio de los clientes

La evaluación de la satisfacción del cliente mostró que:

Las calificaciones promedio son relativamente cercanas entre las tiendas.

Sin embargo, Tienda 2 y Tienda 3 presentan las mejores calificaciones promedio, lo que habla de una experiencia de compra más valorada por los usuarios.

Esto puede influir positivamente en la percepción de los productos nuevos que se introduzcan en esas tiendas.

4. Productos más y menos vendidos

La identificación de productos más vendidos permite ver cuáles tienen mayor tendencia y aceptación.
El análisis mostró que:

Tienda 1 y Tienda 2 tienen mayor cantidad de productos con ventas elevadas.

Tienda 4 muestra más productos con ventas bajas, indicando menor rotación.

Esto sugiere nuevamente que las tiendas con mejor desempeño son la 1 y la 2.

5. Costo de envío promedio por tienda

El análisis del costo de envío (que paga el cliente) mostró:

Diferencias moderadas entre tiendas.

La Tienda 3 y Tienda 4 presentan costos ligeramente más altos.

La Tienda 1 y Tienda 2 tienen costos más estables y competitivos.

Esto es importante porque un envío más económico puede facilitar las ventas, especialmente en productos de bajo precio.

Conclusión y recomendación final

Después de analizar todos los indicadores, la tienda más recomendable para que el Sr. Juan venda sus productos es:

🟩 Tienda 1
Justificación clara:

✔ Mayor ingreso total: demuestra mayor volumen de ventas y flujo comercial.

✔ Buen desempeño en todas las categorías: las categorías más vendidas coinciden con una fuerte rotación.

✔ Productos más vendidos más numerosos: indica mayor capacidad para impulsar nuevos artículos.

✔ Costo de envío competitivo: ayuda a que los clientes finalicen la compra con menos fricción.

Aunque la Tienda 2 también tiene un excelente desempeño (particularmente en calificaciones de clientes), la Tienda 1 presenta el balance más sólido entre ventas, movimiento, rotación de producto y costos operativos, convirtiéndose en la opción más conveniente para maximizar las posibilidades de venta del Sr. Juan.
